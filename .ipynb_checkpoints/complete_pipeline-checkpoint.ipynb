```python
### This script is the complete workflow to get an explanation and diagnosis from eeg data

from preprocessing import preprocess_file
from biomarkers_m import extract_biomarkers
from feature_extraction import extract_features
from ad_prediction import predict
from explanation_model import explain

# setup
path_to_raw_eeg = '/home/jans26/.cache/kagglehub/datasets/rifahramisa/eye-open-eeg-alzheimers/versions/1/eye-open-dataset/sub-001/eeg/sub-001_task-photomark_eeg.set'
biographics = {
    'gender': 'F',
    'age': 57,
    'mmse': 16
}

# preprocess
eeg = preprocess_file(path_to_raw_eeg)

# extract biomarkers
biomarkers = extract_biomarkers(eeg)

# extract eeg feature embedding vector
features = extract_features(eeg)

# compute diagnosis prediction + confidence
prediction, confidence = predict(biographics, biomarkers, features)
print(prediction, confidence)

# explanation
explanation = explain(prediction, confidence, biographics, biomarkers) #TODO Skyler
print(explanation)
```
